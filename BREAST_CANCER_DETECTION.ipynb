```python
from sklearn.model_selection import learning_curve, LearningCurveDisplay
from sklearn.linear_model import LogisticRegression
```

```python
train_sizes, train_scores, test_scores = learning_curve(LogisticRegression(), x, y)
display = LearningCurveDisplay(train_sizes=train_sizes, train_scores=train_scores, test_scores=test_scores, score_name="Score")
display.plot()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test, = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
x_train
```

# Machine Learning Modes

# Logistic Regression

```python
l = LogisticRegression()
l.fit(x_train, y_train)
```

```python
y_pred_lr = l.predict(x_test)
```

```python
y_pred_lr
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
print(accuracy_score(y_test, y_pred_lr))
```

```python
l_acc = accuracy_score(y_test, y_pred_lr)
l_recall = recall_score(y_test, y_pred_lr)
l_precision = precision_score(y_test, y_pred_lr)
print(l_acc)
```

```python
results = pd.DataFrame()
results
```

```python
tempResults = pd.DataFrame({'Algorithm':['Logistic Regresstion Method'], 'Accuracy':[l_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

```python
recall = pd.DataFrame()
tempResults = pd.DataFrame({'Algorithm':['Logistic Regresstion Method'], 'Recall':[l_recall]})
recall = pd.concat( [recall, tempResults] )
recall = recall[['Algorithm','Recall']]
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred_dt = dt.predict(x_test)
y_pred_dt
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred_dt))
```

```python
dt_acc = accuracy_score(y_test, y_pred_dt)
dt_recall = recall_score(y_test, y_pred_dt)
dt_precision = precision_score(y_test, y_pred_dt)
print(dt_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision Tree Classifier'], 'Accuracy':[dt_acc]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision Tree Classifier'], 'Recall':[dt_recall]})
recall = pd.concat( [recall, tempResults] )
recall = recall[['Algorithm','Recall']]
```
